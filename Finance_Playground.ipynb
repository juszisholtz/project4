```python
hist = ticker.history(period="5y", interval="1d")
```

```python
import pandas_ta as ta
hist["RSI"] = ta.rsi(hist["Close"], length=14)
hist["MACD"] = ta.macd(hist["Close"]).iloc[:, 0]
hist["SMA_20"] = hist["Close"].rolling(20).mean()
```

```python
hist['future_return'] = (hist['Close'].shift(-7) - hist['Close']) / hist['Close']
hist['label'] = (hist['future_return'] > 0.05).astype(int)
```

```python
clean_hist = hist.dropna()
clean_hist
```

```python
recs = ticker.recommendations
recs['score'] = recs['To Grade'].map({'Buy': 1, 'Hold': 0, 'Sell': -1}).fillna(0)
recs_agg = recs.resample('D').mean().rolling(7).mean()
clean_hist = clean_hist.merge(recs_agg[['score']], left_index=True, right_index=True, how='left')
```
